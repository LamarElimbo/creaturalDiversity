```python
# Set up data collection foundation

CREATURE_NAME=[]
CREATURE_CAT=[]
SEASON_NUM=[]
EPISODE_NUM=[]
SHOW_NAME=[]

TABLE_COLUMNS = ['creature_name', 'creature_cat', 'season_num', 'show_name', 'episode_num']

DATABASE = {'creature_name':CREATURE_NAME, 'creature_cat':CREATURE_CAT, 'season_num':SEASON_NUM, 'show_name':SHOW_NAME, 'episode_num':EPISODE_NUM}

SUPERNATURAL_URL = 'http://supernatural.wikia.com/wiki/Category:Creatures?page='
BUFFY_URL = 'https://en.wikipedia.org/wiki/List_of_Buffyverse_villains_and_supernatural_beings'
CHARMED_URL = 'http://charmed.wikia.com/wiki/Category:Magical_beings'
GRIMM_URL = 'http://grimm.wikia.com/wiki/Wesen'

seasonLinks = []
creatureLinks = []
counter = 0
```

```python
from bs4 import BeautifulSoup as soup
import pandas as pd
import re
from urllib import request

# Convert html to a more user friendly format
def soupTheLink(url):
    html = request.urlopen(url).read().decode('utf8')
    return soup(html, 'lxml')

# Collect a list of links to the seasons and episodes
def enterEpisode(url):
    souped = soupTheLink(url)

    creatureCategories = ['Demons', 'Vampires', 'Magic Users', 'Enhanced humans', 'Dangerous mortals', 'Undead', 'Ghosts', 'Animalistic beings', 'Deities and higher beings', 'Magical creations', 'Monsters', 'Robots', 'Immortals and non-humans']
    creatureName = ''
    
    body = souped.find(id='mw-content-text')
    currentCat = 0
    counter = 0
    for category in body.find_all(class_='div-col columns column-count column-count-2'):
        for creature in category.find_all('li'):
            counter += 1
            print('creature number ', counter)
            creatureName = creature.get_text()
            print('creature: ', creatureName)
            for episode in creature.find_all('a', href = True):
                try:
                    fullLink = 'http://en.wikipedia.org' + episode['href']
                    collectInfo(fullLink, creatureCategories, currentCat, creatureName)
                except TypeError:
                    pass
        currentCat +=1
```

```python
def collectInfo(url, categoryList, currentCat, creatName):
    souped = soupTheLink(url)

    body = souped.find(id='mw-content-text')
    table = body.find(class_='infobox vevent')
    
    if table == None:
        DATABASE['show_name'].append('Buffy')
        DATABASE['creature_name'].append(creatName)
        DATABASE['creature_cat'].append(categoryList[currentCat])
        DATABASE['season_num'].append('None')
        DATABASE['episode_num'].append('None')
    else:
        for episode in table.find_all('tr'):
          
            prodCode = 0
            for episodeInfo in episode.find_all(['th', 'td']):
                episodeText = episodeInfo.get_text()
                episodeText = episodeText[:15]
                if episodeText == 'Production code':
                    prodCode = 1
                    productionCode = episodeInfo.find_next_sibling()
                    codeCheck = productionCode.find('li')
                    if codeCheck == None:
                        code = episodeInfo.find_next_sibling('td')
                        code = code.get_text()
                        DATABASE['show_name'].append('Buffy')
                        DATABASE['creature_name'].append(creatName)
                        DATABASE['creature_cat'].append(categoryList[currentCat])
                        if productionCode == None:
                            DATABASE['season_num'].append('None')
                            DATABASE['episode_num'].append('None')
                        else:
                            DATABASE['season_num'].append(code[0])
                            DATABASE['episode_num'].append(code[:-2])


                    else:
                        for code in productionCode.find_all(['li']):
                            code = code.get_text()
                            DATABASE['show_name'].append('Buffy')
                            DATABASE['creature_name'].append(creatName)
                            DATABASE['creature_cat'].append(categoryList[currentCat])
                            if productionCode == None:
                                DATABASE['season_num'].append('None')
                                DATABASE['episode_num'].append('None')
                            else:
                                DATABASE['season_num'].append(code[0])
                                DATABASE['episode_num'].append(code[:-2])
                    break
                else:
                    continue
            if prodCode == 1:
                break
```
